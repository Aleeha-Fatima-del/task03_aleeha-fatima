```python
import pandas as pd
```

```python
file_path = '/content/Dataset for Data Analytics (3).xlsx'
df = pd.read_excel(file_path)
display(df.head())
```

```python
import sqlite3
```

```python
# SQLite database creation
conn = sqlite3.connect("orders.db")

# Insert DataFrame into SQL table
df.to_sql("Orders", conn, if_exists="replace", index=False)
```

### Select first 5 rows

```python
display(pd.read_sql("SELECT * FROM Orders LIMIT 5;", conn))
```

### Delivered orders

```python
display(pd.read_sql("SELECT OrderID, Product, TotalPrice FROM Orders WHERE OrderStatus='Delivered';", conn))
```

### Sort by TotalPrice

```python
display(pd.read_sql("SELECT Product, TotalPrice FROM Orders ORDER BY TotalPrice DESC LIMIT 10;", conn))
```

### Revenue per product

```python
display(pd.read_sql("""
SELECT Product, COUNT(*) AS TotalOrders, SUM(TotalPrice) AS Revenue, AVG(UnitPrice) AS AvgPrice
FROM Orders
GROUP BY Product
ORDER BY Revenue DESC;
""", conn))
```

### Coupon usage

```python
display(pd.read_sql("""
SELECT CouponCode, COUNT(*) AS UsageCount
FROM Orders
GROUP BY CouponCode
ORDER BY UsageCount DESC;
""", conn))
```

### Products with revenue > $2000

```python
display(pd.read_sql("""
SELECT Product, SUM(TotalPrice) AS Revenue
FROM Orders
GROUP BY Product
HAVING Revenue > 2000;
""", conn))
```

### Payment method analysis

```python
display(pd.read_sql("""
SELECT PaymentMethod, COUNT(*) AS TotalOrders
FROM Orders
GROUP BY PaymentMethod
ORDER BY TotalOrders DESC;
""", conn))
```

```python
print('--- Checking for duplicate OrderIDs ---')
display(pd.read_sql("SELECT OrderID, COUNT(*) as Count FROM Orders GROUP BY OrderID HAVING Count > 1;", conn))

print('\n--- Sampling Date format ---')
display(pd.read_sql("SELECT Date FROM Orders LIMIT 10;", conn))
```
